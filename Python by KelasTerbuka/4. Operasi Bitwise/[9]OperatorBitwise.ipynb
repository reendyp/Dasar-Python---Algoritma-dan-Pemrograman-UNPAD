```python
# Preview

'''
Operasi masing-masing bitwise, operasi biner, binary
'''
a = 9
b = 5

c = a | b

print('\n========== OR ==========')
print('nilai : ',a,', binary : ', format(a,'08b'))
print('nilai : ',b,', binary : ', format(b,'08b'))
print('--------------------------------------------(|)')
print('nilai : ',c,', binary : ', format(c,'08b'))

#bitwise and (&)
print('\n========== AND ==========')
c = a&b
print('nilai : ',a,', binary : ', format(a,'08b'))
print('nilai : ',b,', binary : ', format(b,'08b'))
print('--------------------------------------------(&)')
print('nilai : ',c,', binary : ', format(c,'08b'))
#bitwise XOR (^)
print('\n========== XOR ==========')

c = a^b
print('nilai : ',a, 'binary : ', format(a,'08b'))
print('nilai : ',b,', binary : ', format(b,'08b'))
print('--------------------------------------------(^)')
print('nilai : ',c,', binary : ', format(c,'08b'))

#bitwise not(~)
print('\n========== NOT ==========')

c = ~a
print('nilai : ',a, 'binary : ', format(a,'08b'))
print('--------------------------------------------(~)')
print('nilai : ',c,', binary : ', format(c,'08b'))
print('--------------------------------------------(^)')

d = 0b0000001001
e = 0b1111111111

print('nilai :',e^d,', binary : ', format(e^d,'08b'))

# shifting

# shift right >>

c = a >> 1
print("\n===========>>============")
print('nilai :',a,',binary :', format(a,'08b'))
print("\n========================= (>>)")
print('nilai :',c,', binary :', format(c,'08b'))

# shift left <<
c = a << 1
print("\n===========<<============")
print('nilai :',a,',binary :', format(a,'08b'))
print("\n========================= (<<)")
print('nilai :',c,', binary :', format(c,'08b'))
```
